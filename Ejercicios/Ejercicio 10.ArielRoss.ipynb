Generar muestras e histograma de las funciones: triangular(), betavariate(), expovariate(), gammavariate(), lognormvariate(), normalvariate(), vonmisesvariate(), paretovariate() y weibullvariate()

```python
import random 
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 4))
plt.hist([random.triangular(0, 1) for i in range(10000)],color="g", bins = 100)
plt.title('Triangular')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.betavariate(5, 10) for i in range(10000)],color="peru", bins = 100)
plt.title('Beta')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.expovariate(1.5) for i in range(10000)],color="dodgerblue", bins = 100)
plt.title('Exponencial')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.gammavariate(1,.5) for i in range(10000)],color="m", bins = 100)
plt.title('Gamma')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.lognormvariate(0,.1) for i in range(10000)],color="coral", bins = 100)
plt.title('Lognormal')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.normalvariate(0,1) for i in range(10000)],color="c", bins = 100)
plt.title('Normal')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.vonmisesvariate(0,1) for i in range(10000)],color="k", bins = 100)
plt.title('Von Mises')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.paretovariate(10) for i in range(10000)],color="r", bins = 100)
plt.title('Pareto')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.hist([random.weibullvariate(1,2) for i in range(10000)],color="y", bins = 100)
plt.title('Weibull')
plt.show()
```
