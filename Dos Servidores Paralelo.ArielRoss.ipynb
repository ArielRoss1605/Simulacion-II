```python
from random import expovariate,random
from math import inf as Infinity
llegadas=2
servidor_a=0.9
servidor_b=0.8
T=100
t=0
clientes=0
t_llegada=expovariate(llegadas)
t_salida_a=Infinity
t_salida_b=Infinity
print("    Llegada--------Clientes---------------------Salida--del--Servidor------------Clientes")
while min(t_llegada,t_salida_a,t_salida_b)<T:
    t=min(t_llegada,t_salida_a,t_salida_b)
    if t==t_llegada:
        clientes=clientes+1
        print("%  10.3f    "%t,"      ",clientes)
        t_llegada=t_llegada+expovariate(llegadas)
        if clientes==1:
            if random()<0.5:
                t_salida_a=t+expovariate(servidor_a)
            else:
                t_salida_b=t+expovariate(servidor_b)
        if clientes==2:
            if t_salida_a<Infinity:
                t_salida_b=t+expovariate(servidor_b)
            else:
                t_salida_a=t+expovariate(servidor_a)
    elif t==t_salida_a:
        clientes=clientes-1
        print("                                           %10.3f   "%t,"      a                  ",clientes)
        if clientes>1:
            t_salida_a=t+expovariate(servidor_a)
        else:
            t_salida_a=Infinity
    else: #t==t_salida_b
        clientes=clientes-1
        print("                                           %10.3f   "%t,"      b                  ",clientes)
        if clientes>1:
            t_salida_b=t+expovariate(servidor_b)
        else:
            t_salida_b=Infinity
        

```
