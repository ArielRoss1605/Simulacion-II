##Datos "en vivo".

A continuaci√≥n veremos diferentes funciones de la libreria investing.

Esta funci√≥n utilizar√° el motor de b√∫squeda Investing.com para recuperar los resultados de b√∫squeda del texto introducido.

```python
buscar_resultado = investpy.search_quotes(text='tesla', products=['stocks'],
                                       countries=['united states'], n_results=1)
print(buscar_resultado)
```

```python
#Recuperaci√≥n de datos recientes.

datos_recientes = buscar_resultado.retrieve_recent_data()

print(f"Datos recientes\n",datos_recientes)

```

```python
#Recuperaci√≥n de datos historicos

datos_historicos = buscar_resultado.retrieve_historical_data(from_date='01/01/2019', to_date='01/01/2020')

print(datos_historicos)
```

```python
#Recuperando informacion del d√≠a de la empresa

informacion = buscar_resultado.retrieve_information()

print(informacion)
```

```python
#Recuperacion de la moneda por default que maneja la empresa 

moneda = buscar_resultado.retrieve_currency()

print(moneda)
```

Un indicador t√©cnico es una variable matem√°tica construida a partir de los datos hist√≥ricos de un activo financiero para hacer √©nfasis en un aspecto particular. Este refleja la trayectoria pasada de este activo, y as√≠ intenta determinar su comportamiento futuro.

```python
#Recuperacion de indicadores tecnicos.

indicadores = buscar_resultado.retrieve_technical_indicators(interval='daily')

print(indicadores)

```

Proporcionan indicadores de compra/venta,analisis de divergencia,estados de sobreventa o sobrecompra,posicion en la que se encuentra,volatilidad,rentabilidad, entre otros.
