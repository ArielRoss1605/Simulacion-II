##Datos "en vivo".

A continuación veremos diferentes funciones de la libreria investing.

Esta función utilizará el motor de búsqueda Investing.com para recuperar los resultados de búsqueda del texto introducido.

```python
buscar_resultado = investpy.search_quotes(text='tesla', products=['stocks'],
                                       countries=['united states'], n_results=1)
print(buscar_resultado)
```

```python
#Recuperación de datos recientes.

datos_recientes = buscar_resultado.retrieve_recent_data()

print(f"Datos recientes\n",datos_recientes)

```

```python
#Recuperación de datos historicos

datos_historicos = buscar_resultado.retrieve_historical_data(from_date='01/01/2019', to_date='01/01/2020')

print(datos_historicos)
```

```python
#Recuperando informacion del día de la empresa

informacion = buscar_resultado.retrieve_information()

print(informacion)
```

```python
#Recuperacion de la moneda por default que maneja la empresa 

moneda = buscar_resultado.retrieve_currency()

print(moneda)
```

Un indicador técnico es una variable matemática construida a partir de los datos históricos de un activo financiero para hacer énfasis en un aspecto particular. Este refleja la trayectoria pasada de este activo, y así intenta determinar su comportamiento futuro.

```python
#Recuperacion de indicadores tecnicos.

indicadores = buscar_resultado.retrieve_technical_indicators(interval='daily')

print(indicadores)

```

Proporcionan indicadores de compra/venta,analisis de divergencia,estados de sobreventa o sobrecompra,posicion en la que se encuentra,volatilidad,rentabilidad, entre otros.

Esta función recupera los valores de las medias móviles calculados por Investing.com para cada producto financiero disponible (acciones, fondos, etfs, índices, cruces de divisas, bonos, certificados y materias primas) para diferentes intervalos de tiempo. Así sucesivamente, el usuario debe proporcionar el nombre product_type y el nombre del producto (a menos que product_type sea 'stock', cuyo valor de nombre será el símbolo de la acción) y el país si es necesario (obligatorio a menos que product_type sea 'currency_cross' o 'commodity', donde debe ser None). Además, se puede especificar el intervalo que define la frecuencia de actualización de los cálculos de las medias móviles (tanto simples como exponenciales). Tenga en cuenta que el intervalo especificado no es el intervalo de la media móvil, ya que se recuperan todos los marcos de tiempo disponibles utilizados en el cálculo de las medias móviles.

```python
#investpy.technical.moving_averages(nombre, país, product_type, intervalo='diario')

investpy.technical.moving_averages('bbva','spain', 'stock','daily')
```
