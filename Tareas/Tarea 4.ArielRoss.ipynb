Simular la aguja de Buffon que se ve en el experimento:

```python
import numpy as np
import matplotlib.pyplot as ptl 
ptl.plot([[0,1],[0,1]],[[-1,-1],[2,2]])#graficamos las lineas
ptl.xlim(-1,2) #los limites del eje x
ptl.ylim(-1,2) #los limites del eje y 
cruce=0 #un contador cuando rebase la linea 
alfiler=int(input("cuantos alfileres quieres arrojar: ")) #cuantos vamos a arrojar 
for i in range(alfiler):
    tetha=np.random.uniform(-np.pi/2,np.pi/2) #tomamos un numero aleatorio para el angulo tetha
    coor=np.random.rand(2) #tomamos un par de puntos aleatorios 
    linf=-0.5*np.cos(tetha)+coor[0]
    lsup=0.5*np.cos(tetha)+coor[0]
    if linf<=0 or lsup>=1: #si esta condicion se cumple entonces...
        cruce=cruce+1 #se genera un cruce 
        ptl.scatter(coor[0],coor[1]) #graficamos el par de puntos aleatorios (coordenadas)
        angulo=np.array([np.cos(tetha),np.sin(tetha)]) #tomamos el angulo del alfiler 
        coorx=coor-0.5*angulo
        coory=coor+0.5*angulo
        ptl.plot(np.array([coorx[0],coory[0]]),np.array([coorx[1],coory[1]])) #graficamos el alfiler y su trayectoria 
print(f'La proximacion de pi es: {2*alfiler/cruce}') #entre mas alfileres mas se aproxima
print(f'Error porcentual: {(np.pi-2*alfiler/cruce)*100/np.pi} %')
```
